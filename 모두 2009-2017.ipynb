```python
import pandas as pd
import plotly
plotly.offline.init_notebook_mode(connected=True)
import plotly.offline as py
import numpy as np

df = pd.read_csv('desktop/data.csv')

for col in df.columns:
    df[col] = df[col].astype(str)
    
df['text'] = df['state'] + '<br>' +\
    'H1B: '+df['h1b']+ '<br>'+\
    'Unemployment: '+df['unemployment']+'<br>'+\
    'GDP: '+df['gdp']+'<br>'+\
    'Greencard: '+df['greencard']
    

data = [dict(type='choropleth',
             z=df['year'].astype(float),
             locations = df['code'].astype(str),
             locationmode='USA-states',
             text = df['text']
            )]


# let's create some additional, random data
for i in range(8):
    data.append(data[0].copy())
    data[-1]['z'] = data[0]['z'] * np.random.rand(*data[0]['z'].shape)

# let's create the steps for the slider
steps = []
for i in range(len(data)):
    step = dict(method='restyle',
                args=['visible', [False] * len(data)],
                label='Year {}'.format(i + 2009))
    step['args'][1][i] = True
    steps.append(step)

sliders = [dict(active=0,
                pad={"t": 1},
                steps=steps)]

layout = dict(title = 'US Immigration Rate by state',
              geo=dict(scope='usa',
                       projection=dict( type='albers usa' ),
                       showlakes = True),
              sliders=sliders)
            

fig = dict(data=data, layout=layout)
plotly.offline.iplot(fig, filename='d3-cloropleth-map')
```
